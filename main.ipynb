```python
# Task 02
#Perform data cleaning and EDA on a dataset od titanic. Expolre the realtionship between variables and
#Identity patterns and trends in a data.

import pandas as pd

#load data of titanic
gender_data=pd.read_csv('gender_submission.csv')
train_data=pd.read_csv('train.csv')
test_data=pd.read_csv('test.csv')


# Display the first few rows of each DataFrame
print('Gender data:')
gender_data.head()
print('Train data:')
train_data.head()
print('Test data:')
test_data.head()
gender_data.info()
train_data.info()
test_data.info()

# Data Cleaning

# Find missing values
missing_value1=gender_data.isnull().sum()
missing_value2=train_data.isnull().sum()
missing_value3=test_data.isnull().sum()

#print missing values
print(missing_value1)
print(missing_value2)
print(missing_value3)

# Drop rows with missing Fare in the test dataset
test_data.dropna(subset=['Fare'], inplace=True)

#Fill missing values
train_data['Age'].fillna(train_data['Age'].median(), inplace=True)
train_data['Embarked'].fillna(test_data['Embarked'].mode()[0], inplace=True)

# Convert categorical columns to numeric
train_data['Sex']=train_data['Sex'].map({'male':0, 'Female':1})
test_data['Sex']=test_data['Sex'].map({'male':0, 'Female':1})

# Encode Embarked column
train_data['Embarked']=train_data['Embarked'].map({'C':0, 'Q':1,'S':2})

# Drop non-numeric columns for numerical analysis
coltodrop=['Name','Cabin','Ticket']
train_data_clean=train_data.drop(columns=coltodrop)
test_data_clean=test_data.drop(columns=coltodrop)

#Display Cleaned Data
print(train_data_clean)
print(test_data_clean)

import matplotlib.pyplot as plt
import seaborn as sns

# Exploratory Data Analysis (EDA)

# Distribution of Survival
plt.figure(figsize=(8,6))
plt.title('Survival Distribution')
sns.countplot(x='Survived', data=train_data_clean)
plt.show()

# Distribution of Age
plt.figure(figsize=(8, 6))
plt.title('Age Distribution')
sns.histplot(train_data_clean['Age'], kde=True, bins=30)
plt.show()

# Survival by Gender
plt.figure(figsize=(8, 6))
sns.barplot(x='Sex', y='Survived', data=train_data_clean)
plt.title('Survival by Gender')
plt.xticks([0, 1], ['Male', 'Female'])
plt.show()
# Survival by Pclass
plt.figure(figsize=(8, 6))
plt.title('Survival by Passenger Class')
sns.barplot(x='Pclass', y='Survived', data=train_data_clean)
plt.show()

# Age vs. Fare
plt.figure(figsize=(10, 6))
plt.title('Age vs. Fare')
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=train_data_clean, palette='viridis')
plt.show()

#Correlation Matrix
plt.figure(figsize=(12, 8))
corr = train_data_clean.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()

# Pairplot to visualize relationships
sns.pairplot(train_data_clean[['Age', 'Fare', 'Pclass', 'Survived']])
plt.title('Pairplot of Features')
plt.show()
```
